## Gray Level Co-Occurrence Matrix (GLCM)
A gray level co-occurrence Matrix is a matrix that is defined over an image to be the distribution of co-occurring pixel values at a given offset. It is used as an approach to texture analysis with various computer vision applications.

```python
patternImage = np.array([
    [0, 0, 1, 1],
    [0, 0, 1, 1],
    [0, 2, 2, 2],
    [0, 2, 3, 3]], dtype=np.uint8)

plt.imshow(patternImage, cmap='gray')
plt.show()
```

We will use <a href="https://scikit-image.org/docs/dev/api/skimage.feature.html?highlight=lbp#skimage.feature.greycomatrix">greycomatrix</a> function from scikit-image to extract GLCM from the sample pattern.

```python
co_ocmatrix = graycomatrix(patternImage, distances=[1], angles=[0, np.pi/4, np.pi/2], levels=4)
print(co_ocmatrix.shape)
print('angle 0')
print(co_ocmatrix[:,:,0,0])

print('angle 45')
print(co_ocmatrix[:,:,0,1])

print('angle 90')
print(co_ocmatrix[:,:,0,2])
```

### GLCM texture properties
Fron the  GLCM we can compute the following texture properties by using <a href='https://scikit-image.org/docs/dev/api/skimage.feature.html?highlight=lbp#skimage.feature.graycoprops'>graycoprops</a> from scikit-image
- contrast $\sum_{i,j=0}^{levels-1} P_{i,j}(i-j)^2$
- dissimilarity $\sum_{i,j=0}^{levels-1}P_{i,j}|i-j|$
- homogeneity $\sum_{i,j=0}^{levels-1}\frac{P_{i,j}}{1+(i-j)^2}$
- correlation $\sum_{i,j=0}^{levels-1} P_{i,j}\left[\frac{(i-\mu_i)(j-\mu_j)}{\sqrt{(\sigma_i^2)(\sigma_j^2)}}\right]$

```python
contrast = graycoprops(co_ocmatrix, prop='contrast')
dissimilarity = graycoprops(co_ocmatrix, prop='dissimilarity')
homogeneity = graycoprops(co_ocmatrix, prop='homogeneity')

print(contrast, dissimilarity,homogeneity)
```

```python
patternImage2 = np.zeros((4,4), dtype=np.uint8) ## all black image!
co_ocmatrix2 = graycomatrix(patternImage2, distances=[1], angles=[0, np.pi/4, np.pi/2])

contrast2 = graycoprops(co_ocmatrix2, prop='contrast')
dissimilarity2 = graycoprops(co_ocmatrix2, prop='dissimilarity')
homogeneity2 = graycoprops(co_ocmatrix2, prop='homogeneity')

print(contrast2, dissimilarity2, homogeneity2)
```

## Local Binary Pattern (LBP)
Local Binary Pattern (LBP) is a efficient texture operator which labels the pixels of an image by thresholding the neighborhood of each pixel and considers the result as a binary number.  It can be seen as a unifying approach to the traditionally divergent statistical and structural models of texture analysis. The most important property of the LBP operator in real-world applications is its robustness to monotonic gray-scale changes caused, for example, by illumination variations. Another important property is its computational simplicity, which makes it possible to analyze images in challenging real-time settings.

#### KL divergence is a way of measuring the matching between two distributions 
(More detail can  be read from <a href="https://towardsdatascience.com/light-on-math-machine-learning-intuitive-guide-to-understanding-kl-divergence-2b382ca2b2a8">here</a>)

```python
def kldivergence(p, q):
    filt = np.logical_and(p != 0, q != 0)
    return np.sum(p[filt] * np.log2(p[filt] / q[filt]))

print('KL-Divergence between crack-crack', kldivergence(hist_crackpatch1 , hist_crackpatch2))
print('KL-Divergence between crack-normal', kldivergence(hist_crackpatch1 , hist_normalpatch1))
```

```python
from scipy.spatial.distance import euclidean

def normalize_patch(patch):
    """Normalize patch to have values between 0 and 31."""
    patch_norm = ((patch - patch.min()) * 31 / (patch.max() - patch.min())).astype(np.uint8)
    return patch_norm

def extract_lbp_features(patch):
    """Extract both LBP and GLCM features optimized for leopard spots."""
    # LBP with larger radius to capture spot patterns
    lbp = local_binary_pattern(patch, P=12, R=3, method='uniform')
    lbp_hist, _ = np.histogram(lbp, bins=24, density=True)
    
    # Normalize patch for GLCM
    patch_norm = normalize_patch(patch)
    
    # GLCM features to capture spot transitions
    glcm = graycomatrix(patch_norm, distances=[2], angles=[0, np.pi/4, np.pi/2], 
                       levels=32, symmetric=True, normed=True)
    contrast = graycoprops(glcm, 'contrast').mean()
    homogeneity = graycoprops(glcm, 'homogeneity').mean()
    
    # Return combined features
    return np.concatenate([lbp_hist, [contrast, homogeneity]])

def create_masked_image(original, mask):
    """Create an image where non-leopard areas are black."""
    # Create RGB version of the image
    result = cv2.cvtColor(original, cv2.COLOR_GRAY2RGB)
    
    # Set non-leopard areas (where mask is 0) to black
    result[mask == 0] = [0, 0, 0]
    
    return result
```

```python
# Compute LBP features for sample patches
leopard_features = [
    extract_lbp_features(leopardPatch1),
    extract_lbp_features(leopardPatch2),
    extract_lbp_features(leopardPatch3)
]
nonleopard_features = [
    extract_lbp_features(nonleopardPatch1),
    extract_lbp_features(nonleopardPatch2),
    extract_lbp_features(nonleopardPatch3)
]

# Initialize segmented image and confidence map
segmented = np.zeros_like(leopardImage, dtype=np.uint8)
confidence_map = np.zeros_like(leopardImage, dtype=float)
count_map = np.zeros_like(leopardImage, dtype=float)

# Sliding window parameters
window_size = 24
step_size = 12

# Perform segmentation using the sliding window approach
for y in range(0, leopardImage.shape[0] - window_size, step_size):
    for x in range(0, leopardImage.shape[1] - window_size, step_size):
        # Extract patch and compute features
        patch = leopardImage[y:y + window_size, x:x + window_size]
        patch_feature = extract_lbp_features(patch)
        
        # Compute distances to both leopard and non-leopard textures
        leopard_distances = [euclidean(patch_feature, f) for f in leopard_features]
        nonleopard_distances = [euclidean(patch_feature, f) for f in nonleopard_features]
        
        # Compute confidence score
        min_leopard_dist = min(leopard_distances)
        min_nonleopard_dist = min(nonleopard_distances)
        confidence = (min_nonleopard_dist - min_leopard_dist) / (min_nonleopard_dist + min_leopard_dist + 1e-6)
        
        # Update confidence map and count
        confidence_map[y:y + window_size, x:x + window_size] += confidence
        count_map[y:y + window_size, x:x + window_size] += 1
        
        # Update segmentation based on confidence
        if confidence > 0:
            segmented[y:y + window_size, x:x + window_size] = 255
```

```python
# Normalize confidence map
valid_mask = count_map > 0
confidence_map[valid_mask] /= count_map[valid_mask]
confidence_map = (confidence_map - confidence_map.min()) / (confidence_map.max() - confidence_map.min() + 1e-6)

# Create masked image
masked_image = create_masked_image(leopardImage, segmented)

# Display results
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.title("Original Image")
plt.imshow(leopardImage, cmap="gray")

plt.subplot(132)
plt.title("Segmentation Mask")
plt.imshow(segmented, cmap="gray")

plt.subplot(133)
plt.title("Masked Result")
plt.imshow(masked_image)

plt.show()
```
