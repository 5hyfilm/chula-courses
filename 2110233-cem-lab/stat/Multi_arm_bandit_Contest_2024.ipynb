# Instruction

There is a 6-arm bandit (many arms can be toggled at once). The bandit will return only one binary reward for each pull, which is zero or one.
In other words, the returning reward depends on the arms being pulled and there are $2^6$ combinations in total.

You will receive an account that can pulls **1000** times.
Your goal is to maximize the cumulative reward of the **last 950 pulls**. In order to be eligible for the competition you must pull at least 50 times.

Note: the arm probability for each person is randomly assigned and will be different for each student.

```
import requests
import math
import random
import json

# API Configuration
url = 'http://183.90.168.139:5001/update_state'
get_state_url = 'http://183.90.168.139:5001/get_state'
token = '3057271549'  # Replace with your actual token

headers = {"Authorization": token}


# Q-function for storing and retrieving Q-values
class QFunction():
    def __init__(self):
        self.q_values = {}

    def get_q_value(self, state, action):
        if (state, action) not in self.q_values:
            return 0  # Default initial value
        return self.q_values[(state, action)]

    def update_q_value(self, state, action, reward, count):
        """Update Q-value using incremental mean."""
        if (state, action) not in self.q_values:
            self.q_values[(state, action)] = reward
        else:
            self.q_values[(state, action)] = (
                (self.q_values[(state, action)] * (count - 1) + reward) / count
            )


# Helper Functions
def pull_arm(arms, times=1):
    """Pull a specific arm(s) a given number of times."""
    payload = {"times": times, "arms": arms}
    try:
        response = requests.post(url, json=payload, headers=headers)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as e:
        print(f"HTTPError: {e}")
        print(f"Response Content: {response.text}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


def get_state():
    """Get the current state of the account."""
    try:
        response = requests.get(get_state_url, headers=headers)
        response.raise_for_status()
        state = response.json()
        print("Current State Retrieved:", json.dumps(state, indent=2))
        return state
    except requests.exceptions.HTTPError as e:
        print(f"HTTPError: {e}")
        print(f"Response Content: {response.text}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


# Main Function
def main():
    num_arms = 64  # Total number of tweak combinations (2^6)
    total_pulls = 0
    max_pulls = 1000
    q_function = QFunction()  # Store Q-values
    ucb = UpperConfidenceBounds()  # Initialize UCB
    total_rewards = [0] * num_arms
    arm_counts = [0] * num_arms

    # State variable (can be extended if needed)
    state = "default"

    # Periodically fetch state to avoid exceeding pull limit
    while total_pulls < max_pulls:
        print(f"Total pulls: {total_pulls}")

        # Select the best arm using UCB
        actions = list(range(num_arms))  # Possible actions are arms [0, 1, ..., 63]
        best_arm = ucb.select(state, actions, q_function)

        # Pull the selected arm
        response = pull_arm([best_arm], times=1)
        if response and "request_reward" in response:
            reward = response["request_reward"][0]

            # Update Q-function and statistics
            arm_counts[best_arm] += 1
            total_rewards[best_arm] += reward
            q_function.update_q_value(state, best_arm, reward, arm_counts[best_arm])

            total_pulls += 1
        else:
            print("Failed to get a response. Skipping this pull.")

        # Fetch state periodically to verify progress and ensure no over-pulls
        if total_pulls % 100 == 0 or total_pulls == max_pulls:
            current_state = get_state()
            if current_state:
                total_pulls = current_state["state"]["count"]
                print(f"Updated Total Pulls: {total_pulls}")

        # Log progress every 50 pulls
        if total_pulls % 50 == 0:
            print(f"Progress: {total_pulls} pulls completed")
            print(f"Cumulative Reward: {sum(total_rewards)}")

    # Output results
    print("Final Results")
    print("Total Rewards:", total_rewards)
    print("Arm Counts:", arm_counts)
    best_arm = total_rewards.index(max(total_rewards))
    print(f"Best Arm: {best_arm} with reward {total_rewards[best_arm]}")

    # Fetch final state
    state_data = get_state()
    if state_data:
        print("Final State:", json.dumps(state_data, indent=2))


# UCB Implementation
class UpperConfidenceBounds():
    def __init__(self):
        self.total = 0
        self.times_selected = {}

    def select(self, state, actions, qfunction):
        # First execute each action one time
        for action in actions:
            if action not in self.times_selected.keys():
                self.times_selected[action] = 1
                self.total += 1
                return action

        max_actions = []
        max_value = float("-inf")
        for action in actions:
            value = qfunction.get_q_value(state, action) + math.sqrt(
                (2 * math.log(self.total)) / self.times_selected[action]
            )
            if value > max_value:
                max_actions = [action]
                max_value = value
            elif value == max_value:
                max_actions += [action]

        # Choose one randomly if there are ties
        result = random.choice(max_actions)
        self.times_selected[result] = self.times_selected[result] + 1
        self.total += 1
        return result


# Execute
if __name__ == "__main__":
    main()
```
